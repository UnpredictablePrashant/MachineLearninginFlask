```python
from sklearn.linear_model import LinearRegression
import pickle
from pandas import *
import csv
```

```python
with open('data.csv') as f:
    readerobj = csv.reader(f)
    input = []
    output = []
    for row in readerobj:
        input.append([row[0],row[1],row[2]])
        output.append([row[3]])

print(input)
print(output)
```

```python
predictor = LinearRegression(n_jobs=-1)
predictor.fit(X=input, y=output)
pickle.dump(predictor, open('model.pkl','wb'))
```

```python
X_TEST = [[10, 20, 30]]
#Expected output will be based on the formulae a+5b+10c which will be 410
outcome = predictor.predict(X=X_TEST)
print('Outcome : {}'.format(outcome))
```
